```python
weight = list(df['AtomicMass'])
weight = [round(elem) for elem in weight]
```

```python
def electron_count(elem):
    n_electrons = elements.index(elem) + 1
    
    return n_electrons

def neutron_count(elem):
    n_neutrons = weight[elements.index(elem)] - elements.index(elem) -1
    return n_neutrons
    
```

```python
neutron_count('Fe')
```

```python
def level_count(elem, n_electrons):
    n_levels = 1
    n_elec = n_electrons
    while max_elec_in_energy_level(n_levels) < n_elec:
        n_elec -= max_elec_in_energy_level(n_levels)
        n_levels += 1
    return n_levels
    
```

```python
def max_elec_in_energy_level(n):
    return 2*n**2
```

```python
def electron_in_energy_level(n_levels, n_electrons):
    electrons_per_level = []
    electrons_left = n_electrons
    for n in range(n_levels):
        electrons_in_level_n = min(electrons_left, max_elec_in_energy_level(n+1))
        electrons_per_level.append(electrons_in_level_n)
        electrons_left -= electrons_in_level_n
    return electrons_per_level
```

```python
def electron_locations(origin, radius, n_electrons):
    
    full_circle = 360
    arc_length = full_circle/n_electrons
    electron_locations = []
    arcs = 0
    for electron in range(n_electrons):
        x0 = origin[0] + radius*math.sin(math.radians(arcs))
        y0 = origin[1] + radius*math.cos(math.radians(arcs))
        arcs += arc_length
        electron_locations.append([x0, y0])
    return electron_locations
    
```
